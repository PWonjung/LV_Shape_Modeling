```python
import pickle
import copy
import open3d as o3d
data= pickle.load(open(r"D:\IdeaProjects\IITP_synth\LV_Shape_Modeling\ipynb\Journal_LVNET\cuttail_temp_L_tex.pkl", "rb"))
print(data.keys())
keys = list(data.keys())

meshes = []
for index in range(len(keys)-1):
    mesh= o3d.geometry.TriangleMesh()
    mesh.vertices = o3d.utility.Vector3dVector(data[keys[0]])
    mesh.triangles = o3d.utility.Vector3iVector(data[keys[index+1]][:, [0, 2, 1]])    
    mesh.paint_uniform_color([0.2*index, 0.2*index, 0.7])   
    
    meshes.append(mesh)
    
o3d.visualization.draw_geometries(meshes[:2], window_name="Temp Mesh: Lateral Ventricle and Hippocampus") # lateral ventricle and hippocampus
o3d.visualization.draw_geometries(meshes[2:], window_name="Temp Mesh: LV Textured with Thalamus, Caudate, and Opposite LV") # LV textured with thalamus, caudate, and opposite lv
```

```python
from PIL import Image

# 이미지 파일 리스트
image_files = ["1-.png", "1+.png", "2-.png", "2+.png", "3-.png", "3+.png"]


for folder in ["pca_ours", "pca_lddmm", "pca_sw"]:
    processed_images = []
    for f in image_files:
        img = Image.open(rf"C:\Users\fabi\Nextcloud\원정\2025\250221 미카이\figures\{folder}\{f}")
        
        # crop (100픽셀씩 제거)
        width, height = img.size
        crop_box = (850, 100, width - 850, height - 100)
        img_cropped = img.crop(crop_box)
        
        # 좌우 반전
        img_flipped = img_cropped.transpose(Image.FLIP_LEFT_RIGHT)
        
        processed_images.append(img_flipped)

    # 오른쪽으로 붙이기
    # 1) 최대 높이 기준으로 새 이미지 생성
    max_height = max(img.height for img in processed_images)
    total_width = sum(img.width for img in processed_images)
    new_img = Image.new("RGB", (total_width, max_height), (255, 255, 255))  # 배경 흰색

    # 2) paste 시 y_offset 중앙 정렬
    x_offset = 0
    for img in processed_images:
        y_offset = (max_height - img.height) // 2
        new_img.paste(img, (x_offset, y_offset))
        x_offset += img.width


    # 결과 저장
    new_img.save(rf"C:\Users\fabi\Nextcloud\원정\2025\250221 미카이\figures\{folder}\merged1.png")
    print("Merged image saved as merged.png")
```

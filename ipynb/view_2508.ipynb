```python
import k3d
import numpy as np

def draw_bounding_box(center, size_x, size_y, size_z, include_points=True):
    """
    K3D에서 바운딩 박스의 12개 모서리를 정확히 그리는 함수
    Parameters:
        center: 바운딩 박스의 중점 (x, y, z)
        size_x: x축 방향 크기 (전체 길이)
        size_y: y축 방향 크기 (전체 길이)
        size_z: z축 방향 크기 (전체 길이)
        include_points: 점 구름 포함 여부
    """
    # 1. 바운딩 박스 꼭짓점 계산
    cx, cy, cz = center
    half_x, half_y, half_z = size_x / 2, size_y / 2, size_z / 2

    # 8개 꼭짓점 정의 (float32로 명시)
    vertices = np.array([
        [cx - half_x, cy - half_y, cz - half_z],  # 0
        [cx + half_x, cy - half_y, cz - half_z],  # 1
        [cx + half_x, cy + half_y, cz - half_z],  # 2
        [cx - half_x, cy + half_y, cz - half_z],  # 3
        [cx - half_x, cy - half_y, cz + half_z],  # 4
        [cx + half_x, cy - half_y, cz + half_z],  # 5
        [cx + half_x, cy + half_y, cz + half_z],  # 6
        [cx - half_x, cy + half_y, cz + half_z]   # 7
    ], dtype=np.float32)

    # 12개 모서리 인덱스 정의 (uint32로 명시)
    # 각 모서리는 독립적인 선분으로 처리되도록 쌍으로 정의
    indices = np.array([
        [0, 1], [1, 2]#, [2, 3], [3, 0],  # z_min 평면
        # [4, 5], [5, 6], [6, 7], [7, 4],  # z_max 평면
        # [0, 4], [1, 5], [2, 6], [3, 7]   # z_min과 z_max 연결
    ], dtype=np.uint32).ravel()  # K3D는 평평한 인덱스 배열을 요구

    # 2. K3D 플롯 생성
    plot = k3d.plot()

    # 카메라 뷰를 바운딩 박스 크기에 맞게 조정
    max_size = max(size_x, size_y, size_z)
    plot.camera_auto_fit = False
    plot.grid_auto_fit = False
    plot.camera = [
        cx + max_size * 2, cy + max_size * 2, cz + max_size * 2,  # 카메라 위치
        cx, cy, cz,  # 타겟 (중점)
        0, 0, 1      # 업 벡터
    ]

    # 선택적으로 점 구름 추가
    # if include_points:
    #     # 바운딩 박스 내부에 무작위 점 생성
    #     np.random.seed(42)
    #     num_points = 100
    #     points = np.random.rand(num_points, 3)
    #     points = points * np.array([size_x, size_y, size_z]) + \
    #              np.array([cx - half_x, cy - half_y, cz - half_z])
    #     point_cloud = k3d.points(positions=points, point_size=max_size * 0.02, color=0x0000ff)
    #     plot += point_cloud

    # 바운딩 박스 모서리 추가 (빨간색 선)
    lines = k3d.line(
        vertices=vertices,
        indices=indices,
        color=0xff0000,  # 빨간색
        width=max_size * 0.01,  # 크기에 비례한 선 두께
        shader='simple'  # 단순 셰이더로 렌더링 문제 방지
    )
    plot += lines

    # 3. 플롯 표시
    plot.display()

# 예제 실행
center = [0, 0, 0]  # 중점 (x, y, z)
size_x, size_y, size_z = 5, 3, 4  # 각 축의 크기
draw_bounding_box(center, size_x, size_y, size_z, include_points=True)
```

```python
import k3d
import numpy as np

def draw_bounding_box(center, size_x, size_y, size_z, include_points=True):
    """
    K3D에서 바운딩 박스의 12개 모서리를 정확히 그리는 함수
    Parameters:
        center: 바운딩 박스의 중점 (x, y, z)
        size_x: x축 방향 크기 (전체 길이)
        size_y: y축 방향 크기 (전체 길이)
        size_z: z축 방향 크기 (전체 길이)
        include_points: 점 구름 포함 여부
    """
    # 1. 바운딩 박스 꼭짓점 계산
    cx, cy, cz = center
    half_x, half_y, half_z = size_x / 2, size_y / 2, size_z / 2

    # 8개 꼭짓점 정의 (float32)
    vertices = np.array([
        [cx - half_x, cy - half_y, cz - half_z],  # 0
        [cx + half_x, cy - half_y, cz - half_z],  # 1
        [cx + half_x, cy + half_y, cz - half_z],  # 2
        [cx - half_x, cy + half_y, cz - half_z],  # 3
        [cx - half_x, cy - half_y, cz + half_z],  # 4
        [cx + half_x, cy - half_y, cz + half_z],  # 5
        [cx + half_x, cy + half_y, cz + half_z],  # 6
        [cx - half_x, cy + half_y, cz + half_z]   # 7
    ], dtype=np.float32)

    # 12개 모서리 정의
    edges = [
        [0, 1], [1, 2], [2, 3], [3, 0],  # z_min 평면
        [4, 5], [5, 6], [6, 7], [7, 4],  # z_max 평면
        [0, 4], [1, 5], [2, 6], [3, 7]   # z_min과 z_max 연결
    ]

    # 2. K3D 플롯 생성
    plot = k3d.plot()

    # 카메라 설정
    max_size = max(size_x, size_y, size_z)
    plot.camera_auto_fit = False
    plot.grid_auto_fit = False
    plot.camera = [
        cx + max_size * 2, cy + max_size * 2, cz + max_size * 2,  # 카메라 위치
        cx, cy, cz,  # 타겟 (중점)
        0, 0, 1      # 업 벡터
    ]

    # 선택적으로 점 구름 추가
    if include_points:
        np.random.seed(42)
        num_points = 100
        points = np.random.rand(num_points, 3)
        points = points * np.array([size_x, size_y, size_z]) + \
                 np.array([cx - half_x, cy - half_y, cz - half_z])
        point_cloud = k3d.points(positions=points, point_size=max_size * 0.02, color=0x0000ff)
        plot += point_cloud

    # 3. 각 모서리를 독립적인 선분으로 추가
    for edge in edges:
        # 각 모서리에 대해 별도의 k3d.line 생성
        line_vertices = vertices[edge].astype(np.float32)
        line = k3d.line(
            vertices=line_vertices,
            color=0xff0000,  # 빨간색
            width=max_size * 0.01,  # 크기에 비례한 선 두께
            shader='simple'  # 단순 셰이더 사용
        )
        plot += line

    # 4. 플롯 표시
    plot.display()

# 예제 실행
center = [0, 0, 0]  # 중점 (x, y, z)
size_x, size_y, size_z = 5, 3, 4  # 각 축의 크기
draw_bounding_box(center, size_x, size_y, size_z, include_points=True)
```

```python
import numpy as np
import open3d as o3d

def pt_to_pt(pt1, pt2):
  pcd1 = o3d.geometry.PointCloud(points=o3d.utility.Vector3dVector(pt1))
  pcd2= o3d.geometry.PointCloud(points=o3d.utility.Vector3dVector(pt2))
  reg_p2p = o3d.pipelines.registration.registration_icp(  # source, target
		pcd1, pcd2, max_correspondence_distance=10,  # Adjust the distance threshold as needed
		estimation_method=o3d.pipelines.registration.TransformationEstimationPointToPoint(),
		criteria=o3d.pipelines.registration.ICPConvergenceCriteria(max_iteration=1000))
  pcd1 = pcd1.transform(reg_p2p.transformation)
  return np.asarray(pcd1.points)

```

```python
def chamfer_loss_np(P, Q, squared=True):
    """
    P: (N,3) array
    Q: (M,3) array
    """
    # pairwise distances
    # shape (N,M,3) -> (N,M)
    diff = P[:, None, :] - Q[None, :, :]
    D2 = np.sum(diff * diff, axis=2)
    if squared:
        d_p = np.min(D2, axis=1).mean()
        d_q = np.min(D2, axis=0).mean()
        return (d_p + d_q)/2
    else:
        D = np.sqrt(D2 + 1e-12)
        d_p = np.min(D, axis=1).mean()
        d_q = np.min(D, axis=0).mean()
        return (d_p + d_q)/2
    
def hausdorff95(P, Q):
    """
    HD95 between two point clouds without using KDTree.
    P: (N,3), Q: (M,3)
    """
    # pairwise distance (N, M)
    dists = np.linalg.norm(P[:, None, :] - Q[None, :, :], axis=2)
    
    # from P to Q: 각 P에서 가장 가까운 Q까지 거리
    min_PQ = np.min(dists, axis=1)
    # from Q to P: 각 Q에서 가장 가까운 P까지 거리
    min_QP = np.min(dists, axis=0)
    
    # 95th percentile
    hd95 = max(np.percentile(min_PQ, 95), np.percentile(min_QP, 95))
    return hd95
from pytorch3d.loss import (chamfer_distance, mesh_laplacian_smoothing, mesh_normal_consistency, point_mesh_face_distance)
from pytorch3d.structures import Meshes, Pointclouds
from pytorch3d.ops import sample_points_from_meshes
def mesh2pcd(vert, faces, lv_target):
    pred_lv_mesh = Meshes(verts=list( vert), faces=list(faces))
    lv_pointclouds = Pointclouds(points=lv_target)
    lv_point_mesh_dist_loss = point_mesh_face_distance(pred_lv_mesh, lv_pointclouds)
    return lv_point_mesh_dist_loss
```
